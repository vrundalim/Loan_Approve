```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,mean_squared_error,root_mean_squared_error,confusion_matrix
import pickle as pk
```

```python
df= pd.read_csv("loan_approval_dataset.csv")
```

```python
df.isna().sum()
```

```python
df
```

```python
df.keys()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.keys()
```

```python
df['education'].unique()
```

```python
df['education'] = df['education'].apply(lambda x : x.strip())
```

```python
df['education'] = df['education'].replace("Graduate","1").replace("Not Graduate","0").astype(int)
```

```python
df['education'].unique()
```

```python
df['self_employed'].unique()
```

```python
df['self_employed'] = df['self_employed'].apply(lambda x : x.strip())
```

```python
df['self_employed'] = df['self_employed'].replace("Yes","1").replace("No","0").astype(int)
```

```python
df['self_employed'].unique()
```

```python
df['loan_status'].unique()
```

```python
df['loan_status'] = df['loan_status'].apply(lambda x :x.strip())
```

```python
df['loan_status'] = df['loan_status'].replace("Approved","1").replace("Rejected","0").astype(int)
```

```python
df['loan_status'].unique()
```

```python
df['Assets'] = df['residential_assets_value'] + df['commercial_assets_value'] + df['luxury_assets_value'] + df['bank_asset_value']
```

```python
df.drop(columns=['loan_id','residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value'],inplace=True)
```

```python
df
```

```python
x = df.drop(columns="loan_status")
y = df["loan_status"]
```

```python
x.keys()
```

```python
# plt.figure(figsize=(10,8))
# sns.heatmap(x.corr(),annot=True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
scaler = StandardScaler()
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.transform(x_test)
```

# Logistic Regression

```python
model1  = LogisticRegression()
```

```python
model1.fit(scaled_x_train,y_train)
```

```python
y_pred1 = model1.predict(scaled_x_test)
```

```python
accuracyscore = accuracy_score(y_test,y_pred1)
mse = mean_squared_error(y_test,y_pred1)
rms = np.sqrt(mse)
```

```python
print("Accuracy_socre",accuracyscore)
print("Mean_squared,error",mse)
print("Mean_squared,error",rms)
```

```python
# sns.heatmap(confusion_matrix(y_test,y_pred1),annot=True)
```

# Decision Tree

```python
model2  = DecisionTreeClassifier(criterion="entropy")
```

```python
model2.fit(scaled_x_train,y_train)
```

```python
y_pred2 = model2.predict(scaled_x_test)
```

```python
accuracyscore = accuracy_score(y_test,y_pred2)
mse = mean_squared_error(y_test,y_pred2)
rms = np.sqrt(mse)
```

```python
print("Accuracy_socre",accuracyscore)
print("Mean_squared,error",mse)
print("Mean_squared,error",rms)
```

```python
# sns.heatmap(confusion_matrix(y_test,y_pred2),annot=True)
```

# Random Forest

```python
model3 = RandomForestClassifier(n_estimators=100)
```

```python
model3.fit(scaled_x_train,y_train)
```

```python
y_pred3 = model3.predict(scaled_x_test)
```

```python
accuracyscore = accuracy_score(y_test,y_pred3)
mse = mean_squared_error(y_test,y_pred3)
rms = np.sqrt(mse)
```

```python
print("Accuracy_socre",accuracyscore)
print("Mean_squared,error",mse)
print("Mean_squared,error",rms)
```
